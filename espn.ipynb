```python
import pandas as pd

df = pd.read_csv("espn.csv")
df.head()
```

```python
#Rename columns 
df=df.rename(columns={"No":"Not_outs","HS":"Highest_inns_score","BF":"Balls_Faced","SR":"Bating_strike_rate","Mat":"Matchs"})
df.head()
```

```python
# Check Null Values 
df.isnull()
```

```python
df.isnull().any()
```

```python
df[df["Balls_Faced"].isnull()]   #isna/isnull
```

```python
#Replace values 
df["Balls_Faced"]=df["Balls_Faced"].fillna(0)
```

```python
df["Bating_strike_rate"]=df["Bating_strike_rate"].fillna(0)
```

```python
df.iloc[[9]]
```

```python
# Drop duplicates 
df.duplicated()
```

```python
#Check for duplicated players , duplicated()==1 equal is true
df[df["Player"].duplicated()==1]
```

```python
df[df["Player"].isin(["GA Headley (WI)","GS Sobers (WI)","JB Hobbs (ENG)","GC Smith (ICC/SA)"])]
```

```python
df =df.drop_duplicates()
```

```python
df
```

```python
#Split Span into Start and End date and Career length
df["Span"].str.split(pat="-")
```

```python
df["Start_date"] = df["Span"].str.split(pat="-").str[0]
```

```python
df["End_date"] = df["Span"].str.split(pat="-").str[1]
```

```python
df.head()
```

```python
#Drop Span column
df=df.drop(["Span"],axis=1)
```

```python
df.head()
```

```python
#Split up country from players 
df["Player"].str.split(pat="(")
```

```python
df["Country"]=df["Player"].str.split(pat="(").str[1]
df["Country"]
```

```python
df["Country"].str.split(pat=")")
```

```python
df["Country"]=df["Country"].str.split(pat=")").str[0]
```

```python
df.head()
```

```python
df["Player"]=df["Player"].str.split(pat="(").str[0]
df.head()
```

```python
#Change Data Type 
#Check Data type in our Dataset
df.dtypes
```

```python
df["Highest_inns_score"]=df["Highest_inns_score"].str.split(pat="*").str[0]
df["Highest_inns_score"]=df["Highest_inns_score"].astype("int")
```

```python
df=df.astype({"Start_date":"int","End_date":"int"})
df.dtypes
```

```python
df['Career_length']=df["End_date"]-df["Start_date"]
df.head()
```

```python
#Question 1 : Average Career length
df["Career_length"].mean()
```

```python
#Question 2 : Average Bating_strike_rate for players who played over 10 years 

df[df["Career_length"]>10]["Bating_strike_rate"].mean()
```

```python
#Question 3 : Numbers of players who played before 1960 

df[df["Start_date"]<1960]["Player"].count()
```

```python
#Question 4 : Max highest inns by country
df.groupby("Country")["Inns"].max()
```
